# Supervised Learning:-

Supervised Learning is a type of Machine Learning where the data contains input features and correct answers (labels).
The model learns by comparing its predictions with the actual answers.

Data + Labels -> Learn mapping

Examples:
- Spam Detection (Spam / Not Spam)
- House Price Prediction
- Student Result Prediction

# Unsupervised Learning:- 

Unsupervised Learning is a type of Machine Learning where the data has no labels. The model finds hidden patterns or groups on its own.

Data only → Find hidden structure

Examples:
- Customer Segmentation
- Market Basket Analysis
- Grouping similar news articles

# Reinforcement Learning:-

Reinforcement Learning is a type of Machine Learning where the model learns by interacting with an environment using rewards and penalties.

Action → Reward/Penalty → Improve decision

Examples:
- Self-driving cars
- Game-playing AI
- Robot navigation systems

### Comparison Table:-

| Learning Type | Definition | Real-World Example 1 | Real-World Example 2 |
|--------------|------------|----------------------|----------------------|
| Supervised Learning | Data has labels (answers) | Spam Detection | House Price Prediction |
| Unsupervised Learning | No labels, find hidden patterns | Customer Segmentation | Fraud Pattern Detection |
| Reinforcement Learning | Learns using rewards and penalties | Self-Driving Cars | Chess/Video Game AI |

Why Choosing the Correct Type Matters
Choosing the wrong learning type is a common beginner mistake.

For example:
- If you try clustering (unsupervised) on a spam dataset that already has labels, you are using the wrong method.
- If you try supervised learning without labels, the model cannot learn correctly.
